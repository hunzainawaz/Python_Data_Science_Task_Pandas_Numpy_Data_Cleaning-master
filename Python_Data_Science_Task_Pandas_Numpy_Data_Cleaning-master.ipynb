## 01. Python Basic: 

i. Employee Scenario. (Make entry for at least 5 employees) 

    - Take input (Name of employee, Salary, Designation (MT, Associate, Senior Associate, Manager), Peformance (bad, average, good))
    - As per govt policy deduct 2% income tax from the salary. 
    - if Designation is MT or Associate please take input and make entry for medical insurance (basic). 
    - if Designation is Senior Associate or Manager please take input for medical insurance (basic) and car allowance (Rs.20,000) and fuel allowance (Rs, 7,000). 
    - if Salary is more than 5 million deduct income tax of around 5% instead of 2%. 
    - Only if performance is good and designation is Senior Associate then make increment in current salary of about 17%
    - Print all employees & their details such as name, Salary, performance, increment amount, medical insurance (basic or advance), car or fuel allownce (if applicable) and their income tax amount. 

```python
n_emp = int(input('How many employees you want to add '))

all_emp_data = []
for i in range(n_emp):
    d_emp_data = {}
    emp_name = input('please enter employee name : ')
    emp_salary = int(input('Please enter employee salary : '))
    emp_desig = input('Please enter employee designation : ')
    emp_perfor = input('Please enter employee performance : ')
    
    d_emp_data['Name']=emp_name
    
    d_emp_data['Designation'] = emp_desig
    d_emp_data['Performance'] = emp_perfor
    
    if(emp_salary >=5000000):
        d_emp_data['Tax'] = emp_salary*5/100
        emp_salary = emp_salary - emp_salary*5/100
    else:
        d_emp_data['Tax'] = emp_salary*2/100
        emp_salary = emp_salary - emp_salary*2/100
    
    if(emp_desig == 'MT' or emp_desig == 'Associate '):
        emp_med = int(input('Kindly Mentioned medical insurance (basic) '))
        d_emp_data['Medical'] = emp_med
    else:
        emp_med = int(input('Kindly Mentioned medical insurance (basic) '))
        d_emp_data['Medical'] = emp_med
        emp_car = 20000
        d_emp_data['Car_Allowance'] = emp_car
        emp_fuel = 7000
        d_emp_data['Fuel_Allowance'] = emp_fuel
        
    if(emp_perfor=='good' and emp_desig == 'Senior Associate'):
        emp_salary = emp_salary + emp_salary*17/100
    else:
        pass
    d_emp_data['Salary']=emp_salary
    all_emp_data.append(d_emp_data)
    

print(all_emp_data)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data.groupby('date').first()
# extracting month
data["Index Name"]= data["date"].str.replace('T000000','')
a = pd.DatetimeIndex(data['Index Name']).month
a.value_counts()

c = a.value_counts()
#variables for graph
x =  c[0:12].index

y = c[0:12].values
# variable for peek month and 
month_and_val = pd.DataFrame(a.value_counts())

# ploting graph
plt.xlabel("Months")
plt.ylabel("Houses registered for sell")
plt.plot(x, y, 'o', color='black');

# printing output
print(month_and_val[month_and_val['Index Name']==a.value_counts().max()])
print('\nAccording to above result and following graph it is clear that the most houses were registered in month 5 (May)')
```
